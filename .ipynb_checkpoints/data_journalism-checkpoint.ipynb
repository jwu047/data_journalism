```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("assets/data/data.csv")
data.drop("id", 1, inplace=True)
data.head()
```

```python
chart_data = data.set_index("state")
chart_data.index.name = None
chart_data = chart_data[["poverty", "income", "healthcare", "obesity", "smokes"]]
chart_data.head()
```

```python
chart_corr = chart_data.corr()
sns.heatmap(chart_corr, annot=True)
plt.show()
```
